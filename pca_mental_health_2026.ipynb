```
#Copia de los datos
data=df.copy()
```

```
# Se crean dummies para las variables categóricas
data = pd.get_dummies(data, columns=['tipo_caso', 'sexo'], drop_first=True, dtype=int) # Se elimina primera columna para variables con solo 2 opciones

data = pd.get_dummies(data, columns=['reg_seg_social', 'estado_civil', 'escolaridad', 'tipo_sust'], drop_first=False, dtype=int) # No se elimina columna para variables con múltiples opciones

data.head()
```

# 2. Correlaciones: relaciones lineales

* No se normaliza

```
# Correlaciones
correlaciones=data.corr()
correlaciones
```

```
#Correlaciones con la variable de interés
cor_variable_interes=correlaciones.loc['num_intentos']
cor_variable_interes
```

```
# Valor absoluto y ordenamiento descendente
cor_variable_interes.abs().sort_values(ascending=False)
```

```
plt.figure(figsize=(6, 8))
plt.barh(y = data.columns, width=cor_variable_interes)
```
